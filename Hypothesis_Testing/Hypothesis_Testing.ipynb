## **Step 1**: Determining Hypothesis

Assuming both samples are independent of each other

H0=  there is no difference in cutlet diameter = μ1 = μ2

Ha= there is a difference in cutlet diameter  =  μ1  ≠ μ2 


## Step 2 : Determine test stats to be used for test

We will be using **2-tailed t test** for this question

Significance level = 5%

alpha =0.05 / 2               

##  Step 3 : 

```python
data=pd.read_csv('Cutlets.csv')
```

```python
data.head()
```

```python
alpha=0.05/2
```

```python
data.describe()
```

```python
data[data.duplicated()].shape
```

```python
plt.subplots(figsize=(9,6))
plt.subplot(121)
plt.boxplot(data['Unit A'])
plt.title("Unit A")
plt.subplot(122)
plt.boxplot(data['Unit B'])
plt.title('Unit B')
plt.show()
```

```python
plt.subplots(figsize=(7,5))
plt.subplot(121)
plt.hist(data['Unit A'])
plt.title("Unit A")
plt.subplot(122)
plt.hist(data['Unit B'])
plt.title('Unit B')
plt.show()
```

```python
plt.figure(figsize=(6,4))
label=['Unit A','Unit B']
sns.distplot(data['Unit A'],kde=True)
sns.distplot(data['Unit B'],kde=True)
plt.legend(label)
```

```python
plt.figure(figsize=(2,2))
sm.qqplot(data['Unit A'],line='q')
plt.title('Unit A')
sm.qqplot(data['Unit B'],line='q')
plt.title('Unit B')
plt.show()
```

## Step 4 : Calculate p_value and compare it to test stats

```python
stat ,p_value =stats.ttest_ind(data['Unit A'],data['Unit B'])
```

```python
p_value
```

## Step 5 : Decision 

if **pvalue < alpha**,we reject the null hypothesis

or if **p value >alpha**,  we fail to reject null hypothesis

```python
if p_value < alpha:
    print( " We reject null hypothesis since p_value is greater than alpha" )
else:
    print("We failed to reject null hypothesis, because of lack of evidence that two samples are not similar diameter")
```

## Step 2 Determine the test stats
We will be conducting Test of Independence using Chi square test with contigency table

Significance level = 5%

alpha = 0.05

##  Step 3  Applying Chi square contigency value converting observed in expected values

```python
'''
This one works after getting rid of observed value(str) part from 'buyers' and saving it. '''

stat, p, dof, exp = stats.chi2_contingency(buyers)   
```

```python
observed
```

```python
expected
```

```python
stat,p_value=stats.chisquare(observed,expected,ddof=3)
```

```python
p_value
```

## Step 4 Comparing p_value with alpha

```python
if p_value <= alpha:
    print(f'"We reject null hypothesis there is relation between buyer ration based on gender and regions."')
else:
    print(f'"We failed to reject null hypothesis there is no relation between buyer ration based on gender and regions".')
```

**We failed to reject null hypothesis there is no relation between buyer ration based on gender and regions**
